```python
# sort stock price for nvidia to get only 2024 numbers 

datanvda = nvda.history(period="5y").reset_index()

nvda2024=[]
nvda2024_x=[]


for i in range(0,len(datanvda)): 
    if "2020" in str(datanvda["Date"][i]): #str(datanvda["Date"][i])[5:7] == "10" or str(datanvda["Date"][i])[5:7] == "09" or str(datanvda["Date"][i])[5:7] == "11"
        nvda2024.append(datanvda["Close"][i])
        nvda2024_x.append(str(datanvda["Date"][i])[5:10])

    i+=1

data = pd.DataFrame({"Date":nvda2024_x, "Close":nvda2024})

data.to_csv("nvda2024.csv", index=False)

plt.plot(nvda2024_x, nvda2024)
plt.xticks(rotation='vertical')
```
