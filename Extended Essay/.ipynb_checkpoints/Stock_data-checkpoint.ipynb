```python
# S&P 500 stock data (5 year period)

import yfinance as yf
import pandas as pd

# Specify the ticker symbol for the S&P 500
ticker_symbol = "^GSPC"

# Download the data for the last 5 years
data = yf.download(ticker_symbol, period="5y", interval="1d")

# Display the data
print(data)

# Further analysis can be done with pandas
# For example, calculate the daily return:
# data["daily_return"] = data["Close"].pct_change()
# print(data.head())

data['Close'].tail(10)
```

```python
nvda = yf.Ticker("NVDA")
asml = yf.Ticker("ASML")
aapl = yf.Ticker("AAPL")
goog = yf.Ticker("GOOG")
tsmc = yf.Ticker("TSM")
s_p = yf.Ticker("^GSPC")
ndx = yf.Ticker("^NDX")
spgtsi = yf.Ticker("^SPGSTI")



print(nvda.history(period="5y").tail)

data = {
    # "Date": nvda.history(period="5y")['Date'],
    "NVDA": nvda.history(period="5y")['Close'],
    "ASML": asml.history(period="5y")['Close'],
    "AAPL": aapl.history(period="5y")['Close'],
    "GOOG": goog.history(period="5y")['Close'],
    "TSMC": tsmc.history(period="5y")['Close'],
    "S&P_500": s_p.history(period="5y")['Close'],
    "Nasdaq": ndx.history(period="5y")['Close'],
    "spgtsi": spgtsi.history(period="5y")['Close']
}

stock_data = pd.DataFrame(data)

stock_data
```

```python
# sort stock price for nvidia to get only 2024 numbers 

datanvda = nvda.history(period="5y").reset_index()

nvda2024=[]
nvda2024_x=[]


for i in range(0,len(datanvda)): 
    if "2024" in str(datanvda["Date"][i]): #str(datanvda["Date"][i])[5:7] == "10" or str(datanvda["Date"][i])[5:7] == "09" or str(datanvda["Date"][i])[5:7] == "11"
        nvda2024.append(datanvda["Close"][i])
        nvda2024_x.append(str(datanvda["Date"][i])[5:10])

    i+=1

data = pd.DataFrame({"Date":nvda2024_x, "Close":nvda2024})

plt.plot(nvda2024_x, nvda2024)
plt.xticks(rotation='vertical')
```
