```python
import matplotlib.pyplot as plt
from numpy import *
from math import *
```

```python
b=[]
d=[]
u=[]

for i in linspace(0,4*pi,50):
    c=sin(i)
    f=cos(i)
    p=sin(i**2)

    b.append(c)
    d.append(f)
    u.append(p)

x2=linspace(0,20,50)
```

```python
plt.plot(x2,b, label="sin", color="red", marker="*", linestyle=":")

plt.plot(x2,d, label="cos", color="blue", marker="*", linestyle="-.")

plt.plot(x2,u, label="sin(squared)", color="green", marker="*", linestyle="-")

plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("idk, some plot")
```
