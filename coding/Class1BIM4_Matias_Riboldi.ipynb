```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.linspace(0,8*np.pi,301)
b = np.linspace(0,8*np.pi,301)
def Y(x):
    return (2*np.cos(x)+10)

def Y2(x):
    return (4*np.cos(x)+10)

y = []
v = 0

for i in x: 
    if v < 150:
        y.append(Y(i))
    else: 
        y.append(Y2(i))
    v+=1

#  Matias Riboldi
```

```python
plt.plot(x,y, label="y")
plt.plot(x,b, label="b")
plt.title("cosine(amp=2) to cosine(amp=4) function")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
